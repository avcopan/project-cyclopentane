```python
from pathlib import Path

import automech
```

```python
mech_num = "04"
# submech_leaf = "1submech_ene"
# submech_leaf = "2submech_12epoxy"
submech_leaf = "3submech_13epoxy"
# task_leaf = "1generate"
# mech_tag = ""
task_leaf = "2stereoexpand"
mech_tag = "_sorted"

dir_path = Path("..") / submech_leaf / task_leaf
assert dir_path.exists()

mech_path = dir_path / f"mechanism{mech_num}.dat{mech_tag}"
spc_path = dir_path / f"species{mech_num}.csv{mech_tag}"
```

```python
mech = automech.io.mechanalyzer.read.mechanism(
    rxn_inp=mech_path,
    spc_inp=spc_path,
)
mech
```

```python
automech.display_reactions(mech, eqs=["cyC5x13r2rs = C5H6O-HSdVHzr1+H"])
```

```python
import automol

rsmis = ('[CH]1[C@H]2CC[C@@H]1O2',)
psmis = ('C1=C2CC[C@@H]1O2', '[H]')
rxns = automol.reac.from_smiles(rsmis, psmis)
print(len(rxns))
```

```python
rxn, *_ = rxns
ts_gra = automol.reac.ts_graph(rxn)
ts_geo = automol.graph.geometry(ts_gra)
automol.geom.display(ts_geo, gra=ts_gra)
```
