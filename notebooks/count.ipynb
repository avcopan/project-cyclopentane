```python
# Count total number of direct / well-skipping rate coefficients
tag = "Y_reactions_v0"

mech_json = p_.stereo_mechanism(tag, "json", path=p_.data(root_path))
mech = am.io.read(mech_json)
mech = am.merge_resonant_instabilities(mech, remove=True)

# Drop barrierless reactions
ramchi_col = "reactant_amchis"
pamchi_col = "product_amchis"
spc_df = mech.species
rxn_df = mech.reactions.filter(~pl.col("reactants").list.contains("O2(6)"))
rxn_df = am.reaction.translate_reagents(
    rxn_df,
    trans=spc_df["name"],
    trans_into=spc_df["amchi"],
    rct_col=ramchi_col,
    prd_col=pamchi_col,
)

counts = []
data = list(rxn_df.select("reactants", "products", ramchi_col, pamchi_col).iter_rows())
for num, (rcts, prds, rchis, pchis) in enumerate(data):
    print(f"{num + 1} / {len(data)}")
    print(rcts, prds)
    print(rchis, pchis)
    rxn_objs = automol.reac.from_amchis(rchis, pchis)
    if rxn_objs:
        conn_count = len(rxn_objs)
        print(f"count = {conn_count}")
        automol.reac.display(rxn_objs[0])
        counts.append(conn_count)
    else:
        counts.append(1)

print(f"Total: {sum(counts)}")
```
