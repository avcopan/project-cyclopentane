```python
from pathlib import Path

from IPython import get_ipython

import automech

tag, *_ = Path(get_ipython().user_ns.get("__vsc_ipynb_file__")).stem.split(".")
```

```python
DATA_PATH = Path("../data")
CALC_PATH = Path(f"../{tag}/data/CKIN")
```

```python
# Read mechanism
par_mech0 = automech.io.read(DATA_PATH / "full_raw.json")
mech0 = automech.io.read(DATA_PATH / f"{tag}.json")
```

```python
# Read and update thermo
*_, therm_file = CALC_PATH.glob("all_therm.ckin*")
mech = automech.io.chemkin.update.thermo(mech0, therm_file)
```

```python
# Set rate units to match parent
mech = automech.set_rate_units(mech, automech.rate_units(par_mech0))
```

```python
# Read and update rates
rate_files = list(CALC_PATH.glob("*.ckin"))
mech = automech.io.chemkin.update.rates(mech, rate_files)
```

```python
print(automech.io.chemkin.write.reactions_block(mech))
```
