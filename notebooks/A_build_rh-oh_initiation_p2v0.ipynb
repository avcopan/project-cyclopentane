```python
from pathlib import Path

from IPython import get_ipython

import automech

stem = Path(get_ipython().user_ns.get("__vsc_ipynb_file__")).stem
```

```python
DATA_PATH = Path("../data")
```

```python
# Read
mech0 = automech.io.read(DATA_PATH / "1ene_named.json")
```

```python
# Build
#  - RMG selection: radius 1 neighborhood of *ene* on RH + OH PES
mech = automech.select_pes(mech0, "C5H9O")
mech = automech.neighborhood(mech, ["C5H8(522)"], radius=1)
mech = automech.without_duplicate_reactions(mech)
```

```python
# Expand and sort
mech, err_mech = automech.expand_stereo(mech, distinct_ts=False)
mech = automech.with_sort_data(mech)
```

```python
# Write
automech.io.write(mech, DATA_PATH / f"{stem}.json")
automech.io.mechanalyzer.write.mechanism(
    mech,
    rxn_out=DATA_PATH / "mechanalyzer" / f"{stem}.dat",
    spc_out=DATA_PATH / "mechanalyzer" / f"{stem}.csv",
)
```
