```python
from pathlib import Path

import automech
```

```python
DATA_PATH = Path("data")
```

```python
par_tag = "full_cyclopentane"
sub_tag = "cyclopentene"
tag = "1ene"
```

```python
par_mech = automech.io.rmg.read.mechanism(
    rxn_inp=DATA_PATH / "rmg" / f"{par_tag}_chem.inp",
    spc_inp=DATA_PATH / "rmg" / f"{par_tag}_species_dictionary.txt",
)
sub_mech = automech.io.rmg.read.mechanism(
    rxn_inp=DATA_PATH / "rmg" / f"{sub_tag}_chem.inp",
    spc_inp=DATA_PATH / "rmg" / f"{sub_tag}_species_dictionary.txt",
)
name_dct, missing_names = automech.rename_dict(sub_mech, par_mech)
print(missing_names)
```

```python
sub_mech = automech.rename(sub_mech, name_dct)
automech.io.write(sub_mech, DATA_PATH / f"00_{tag}_orig.json")
```

```python
sub_mech
```

```python
exp_mech, err_mech = automech.expand_stereo(sub_mech)
```

```python
automech.io.write(exp_mech, DATA_PATH / f"00_{tag}_exp.json")
automech.io.write(err_mech, DATA_PATH / f"00_{tag}_err.json")
```
