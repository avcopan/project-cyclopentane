```python
from pathlib import Path

import automech
```

```python
mech = automech.io.read(Path("data") / "00_1ene_orig.json")
mech = automech.pes_filter(mech, exclude_formulas=["C5H8O2"])
net = automech.network(mech)
```

```python
sub_net = automech.net.neighborhood(
    net, species_names=["C5H7(1)", "C5H7(2)", "C5H7(3)"], radius=1
)
automech.net.display(sub_net, height="950px", out_name="net1.html")
```
