```python
class A:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def __repr__(self):
        return f"This is an instance of A with values a={self.a} and b={self.b}"
```

```python
a_obj = A(5, 6)
print(a_obj)
```
