```python
import automech
```

```python
mech = automech.from_smiles(
    rxn_smis=[
        "CC.[OH]>>C[CH2].O",
        "C[CH2].[O][O]>>CCO[O]",
        "C[CH2]>>C=C.[H]",
        "CCO[O]>>C=C.O[O]",
        "CCO[O]>>[CH2]COO",
        "CCO[O]>>C[CH]OO",
    ]
)

mech = automech.with_sort_data(mech)

rxn_str, spc_str = automech.io.mechanalyzer.write.mechanism(mech)
print(rxn_str)
```
