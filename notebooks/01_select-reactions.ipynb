```python
from pathlib import Path

import automech
```

```python
DATA_PATH = Path("data")
tag = "1ene"
sub_tag = "nei1"
```

```python
mech = automech.io.read(DATA_PATH / f"00_{tag}_orig.json")
net = automech.network(mech)
```

```python
sub_net = automech.net.neighborhood(net, ["C5H7(1)", "C5H7(2)", "C5H7(3)"])
sub_mech = automech.from_network(sub_net)
```

```python
exp_mech, err_mech = automech.expand_stereo(sub_mech)
```

```python
automech.io.write(exp_mech, DATA_PATH / f"01_{tag}_{sub_tag}_exp.json")
automech.io.write(err_mech, DATA_PATH / f"01_{tag}_{sub_tag}_err.json")
```

```python
exp_mech = automech.without_duplicate_reactions(exp_mech)
automech.io.mechanalyzer.write.mechanism(
    exp_mech,
    rxn_out=DATA_PATH / "mechanalyzer" / f"01_{tag}_{sub_tag}_exp_mechanism.dat",
    spc_out=DATA_PATH / "mechanalyzer" / f"01_{tag}_{sub_tag}_exp_species.csv",
)
```
