```python
from pathlib import Path

import automech
```

```python
mech_num = "07"
submech_leaf = "1submech_ene"
# submech_leaf = "2submech_12epoxy"
# submech_leaf = "3submech_13epoxy"
inp_task_leaf = "1generate"
inp_mech_tag = ""
out_task_leaf = "2stereoexpand"
out_mech_tag = "_new"

inp_path = Path("..") / submech_leaf / inp_task_leaf
assert inp_path.exists()
out_path = Path("..") / submech_leaf / out_task_leaf
out_path.mkdir(exist_ok=True)

rxn_inp = inp_path / f"mechanism{mech_num}.dat{inp_mech_tag}"
spc_inp = inp_path / f"species{mech_num}.csv{inp_mech_tag}"
rxn_out = out_path / f"mechanism{mech_num}.dat{out_mech_tag}"
spc_out = out_path / f"species{mech_num}.csv{out_mech_tag}"
```

```python
mech = automech.io.mechanalyzer.read.mechanism(rxn_inp=rxn_inp, spc_inp=spc_inp)
mech, err_mech = automech.expand_stereo(mech)
automech.io.mechanalyzer.write.mechanism(mech, rxn_out=rxn_out, spc_out=spc_out)
print(mech)
```

```python
automech.display(mech)
automech.display_reactions(err_mech)
```
