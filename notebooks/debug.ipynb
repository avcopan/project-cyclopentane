```python
import automol

rsmis = ("[O]OC1CCC2C1O2",)
psmis = ("OOC1[CH]CC2C1O2",)

rxn, *_ = automol.reac.from_smiles(rsmis, psmis, stereo=False)
```

```python
automol.smiles.inchi("OO[C@@H]1CC=C[CH]1")
```

```python
ts_gra = automol.reac.ts_graph(rxn)
automol.graph.display(ts_gra)
```

```python
print(ts_gra)
```

```python
ts_sgras = automol.graph.expand_stereo(ts_gra, enant=False)
for ts_sgra in ts_sgras:
    automol.graph.display(ts_sgra)
```
