```python
from pathlib import Path

import automech
```

```python
DATA_PATH = Path("data")
```

```python
tags = ["1ene", "2epoxy12"]
```

```python
par_mech = automech.io.read(DATA_PATH / "full_raw.json")

for tag in tags:
    mech = automech.io.read(DATA_PATH / f"{tag}_raw.json")
    name_dct, missing_names = automech.rename_dict(mech, par_mech)
    print(missing_names)
    mech = automech.rename(mech, name_dct)
    automech.io.write(mech, DATA_PATH / f"{tag}_named.json")
    print(mech)
```
