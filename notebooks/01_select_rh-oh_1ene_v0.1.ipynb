```python
from pathlib import Path

import automech
```

```python
DATA_PATH = Path("data")
tag = "1ene"
radicals = ["C5H7(1202)", "C5H7(500)", "C5H7(504)"]
```

```python
mech = automech.io.read(DATA_PATH / f"{tag}_named.json")
net = automech.network(mech)
```

```python
sub_net = automech.net.neighborhood(net, radicals)
automech.net.display(sub_net)
```
