```python
inp = 'cyclopentene_chem.inp'
spc_inp = 'cyclopentene_species.csv'
inp = '12epoxycyclopentane_chem.inp'
spc_inp = '12epoxycyclopentane_species.csv'
inp = '13epoxyCPT_chem.inp'
spc_inp = '13epoxyCPT_species.csv'

out = inp.replace('_chem.inp', '_reactions.csv')
all_out = inp.replace('_chem.inp', '_reactions_all.csv')
err_out = inp.replace('_chem.inp', '_reactions_unclassified.csv')
```
