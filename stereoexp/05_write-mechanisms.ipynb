```python
import mecha
import pandas
from mechanalyzer.builder import sort_fct as ma_sort
from mechanalyzer.parser import spc as ma_spc_io
import chemkin_io

from IPython import display as ipd
```

```python
def write_mechanisms(rxn_path, spc_path, drop=True):
    rxn_df = pandas.read_csv(rxn_path)
    spc_df = pandas.read_csv(spc_path)

    mech_str, spc_str = mecha.to_mechanalyzer(rxn_df, spc_df, drop=drop)

    mech_path = rxn_path.replace('.csv', '_mech.ckin')    
    with open(mech_path, 'w') as f:
        f.write(mech_str)

    spc_path = spc_path.replace('.csv', '_mech.csv')
    with open(spc_path, 'w') as f:
        f.write(spc_str)

    return mech_str, spc_str
```

```python
mech_str, spc_str = write_mechanisms(
    rxn_path='cyclopentene_reactions_stereo.csv',
    spc_path='cyclopentene_species_stereo.csv',
)
print(spc_str)
print("---")
print(mech_str)
```
