```python
from pathlib import Path
import automol
import numpy as np
```

```python
root_dir0 = Path("/lcrc/project/PACC/copan/proj/project-cyclopentane/save/RXN/C5H9O/ZOCQCVFDVWWLHS/0/2/YFKPBYRVSA-N/C5H8.HO/UGZOASBOAAXKSP/0_0/1_2/UHFFFAOYSA-N/2/u-unGOU/TS/00/CONFS/rj4uSanU6dWtt")
conf_dirs0 = [d for d in root_dir0.iterdir() if d.is_dir()]
print(len(conf_dirs0))

root_dir = Path("confs")
root_dir.mkdir(exist_ok=True)

for conf_dir0 in conf_dirs0:
    geo_file0 = conf_dir0 / "geom.xyz"
    hess_file0 = conf_dir0 / "hess.hess"
    geo = automol.geom.from_xyz_string(geo_file0.read_text())
    gra = automol.geom.graph_without_stereo(geo)
    gra = automol.graph.add_bonds(gra, [(0, 13)], ord_dct={(0, 13): 0.1})

    hess = np.loadtxt(hess_file0)
    freqs, modes = automol.geom.vibrational_analysis(geo, hess)
    print(freqs)
    # automol.geom.display(geo, mode=modes[:, 0])
    automol.geom.display(geo, gra=gra)
    geo_forw = automol.geom.displace(geo, +1.2 * modes[:, 0])
    geo_back = automol.geom.displace(geo, -1.2 * modes[:, 0])
    automol.geom.display(geo_forw)
    automol.geom.display(geo_back)

    conf_dir = root_dir / conf_dir0.name
    conf_dir.mkdir(exist_ok=True)
    geo_forw_file = conf_dir / "forw.xyz"
    geo_back_file = conf_dir / "back.xyz"
    geo_forw_file.write_text(automol.geom.xyz_string(geo_forw))
    geo_back_file.write_text(automol.geom.xyz_string(geo_back))

    zma_forw = automol.geom.zmatrix(geo_forw, gra=gra)
    zma_back = automol.geom.zmatrix(geo_back, gra=gra)
    zma_forw_file = conf_dir / "forw.zmat"
    zma_back_file = conf_dir / "back.zmat"
    zma_forw_file.write_text(automol.zmat.string(zma_forw))
    zma_back_file.write_text(automol.zmat.string(zma_back))
```
