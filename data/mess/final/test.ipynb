```python
from importlib import reload
import functools
import operator

from autochem.rate import data

reload(data)

rate_strs = [
    r"""
W7->W14

P\T           300       400       500       600       700       800       900     1e+03   1.1e+03   1.2e+03
0.1      2.01e-07 -0.000441     -0.17      60.8     -90.7        13  2.87e+03       702  3.54e+03  1.55e+04
1        6.25e-06 -8.58e-06      1.98      -185      -777       580  3.49e+03  1.56e+04  5.97e+04  1.92e+05
10      -2.18e-06  0.000228   -0.0383      2.33       183  3.23e+03  2.52e+04  1.19e+05  4.14e+05  1.22e+06
100      3.25e-07  2.96e-07      1.34      27.6       244  5.73e+03  6.85e+04  3.79e+05  1.52e+06  4.62e+06
O-O       4.5e-13  1.12e-06   0.00849      3.44       257   6.6e+03  8.31e+04  6.32e+05  3.33e+06  1.33e+07
    """
]

rates = [data.from_mess_channel_output(s, order=1) for s in rate_strs]
rate = functools.reduce(operator.add, rates)
rate_fit = data.ArrheniusRateFit.fit(rate.T, rate(T=rate.T, P=0.1), A_fill=1e-20, fill_bad_fit=True)
# rate_fit = data.ArrheniusRateFit.fit(rate.T, rate(T=rate.T, P=0.1))
data.display([rate, rate_fit], label=["data", "fit"], P=0.1, T_range=(300, 1200))
```
