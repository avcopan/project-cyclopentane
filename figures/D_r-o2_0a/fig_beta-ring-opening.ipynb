```python
def print_arrhenius_parameters(rate, name):
    assert isinstance(rate, ac.rate.data.PlogRateFit)
    ix = next(i for i, P in enumerate(rate.Ps) if P == 1)
    k = rate.arrhenius_functions[ix]
    print(f"{name}\t{k.A:.4e}\t{k.b:.4f}\t{k.E:.2f}")

print_arrhenius_parameters(open45_rxn_fit.rate, "S(722)r0 = S(728)e AND S(722)r0 = S(728)z\tRing-opening → 1-yl-4-ene-5-hydroperoxy")
print_arrhenius_parameters(open34_rxn_fit.rate, "S(722)r0 = S(731)r0\tRing-opening → 1-yl-4-ene-3-hydroperoxy")
print_arrhenius_parameters(e12_rxn_fit.rate, "S(722)r0 = C5H8O(825)rs + OH(4)\tRing-closing (1,2) → OH")
print_arrhenius_parameters(sciss_rxn_fit.rate, "S(722)r0 = C5H8(522) + HO2(8)\tBeta scission → HO₂")
```
