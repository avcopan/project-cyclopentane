```python
from pathlib import Path

import automech as am
from project_utilities import p_
from protomech import mess

open_browser = True
tag = "Z_mess_v6"
mech_tag = "Z_mess_v1"
root_path = Path("../..")

stoich = "C5H7"
mess_file_path = p_.mess_final(root_path, tag) / stoich / "mess.inp"
print(mess_file_path)

mech_json = p_.stereo_mechanism(mech_tag, "json", path=p_.data(root_path))
mech = am.io.read(mech_json)
surf = mess.surf.from_mess_input(mess_file_path)
surf = mess.surf.shift_energies(surf, shift=-mess.surf.node_object(surf, 1).energy)
surf = mess.surf.absorb_fake_nodes(surf)

mess.net.display(surf, height="1000px", mech=mech)
```

```python
from autochem.util import plot

root_key = 1
leaf_keys = [2, 7, 6]
color_cycle = [plot.Color.purple, plot.Color.green, plot.Color.red]
color_dct = dict(zip(leaf_keys, color_cycle))

node_paths = mess.surf.node_paths_from_source(surf, root_key, leaf_keys=leaf_keys)
print(node_paths)

colors = [color_dct.get(node_path[-1], plot.Color.black) for node_path in node_paths]

mess.surf.plot_paths(
    surf,
    node_paths,
    stack_nodes=[[0, 3]],
    colors=colors,
    label=True,
)
```
