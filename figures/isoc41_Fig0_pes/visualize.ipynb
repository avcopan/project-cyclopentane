```python
from pathlib import Path

import automech
from protomech import mess

mess_inp = Path("test.inp")
mech_file = Path("test.json")
```

```python
root_key = 2

surf = mess.surf.from_mess_input(mess_inp)
surf = mess.surf.absorb_fake_nodes(surf)
surf = mess.surf.contract_enantiomer_node_pairs(surf)

energy_dct = mess.surf.energy_dict(surf)
surf = mess.surf.shift_energies(surf, -energy_dct[root_key])

mech = automech.io.read(mech_file)
mess.net.display(surf, height="1000px", mech=mech)
```
