```python
from autochem.util import plot

def lighten_hex_colors(
    base_hex: str,
    n: int,
    max_lighten: float = 0.8,
) -> list[str]:
    """
    Generate n progressively lighter shades of a base color.

    Parameters
    ----------
    base_hex : str
        Base color like '#8533FF'.
    n : int
        Number of colors.
    max_lighten : float
        Maximum fraction toward white (0–1). 1 = pure white.

    Returns
    -------
    list[str]
        List of hex color strings.
    """
    base_hex = base_hex.lstrip("#")
    r = int(base_hex[0:2], 16)
    g = int(base_hex[2:4], 16)
    b = int(base_hex[4:6], 16)

    if n <= 1:
        return [f"#{r:02X}{g:02X}{b:02X}"]

    step = max_lighten / (n - 1)

    colors = []
    for i in range(n):
        t = i * step
        r_new = round(r + t * (255 - r))
        g_new = round(g + t * (255 - g))
        b_new = round(b + t * (255 - b))
        colors.append(f"#{r_new:02X}{g_new:02X}{b_new:02X}")

    return colors

root_key = 19
leaf_keys = [16, 17, 18, 15, 14, 21]
color_cycle = [plot.Color.blue, *lighten_hex_colors(plot.Color.purple, n=len(leaf_keys)-1)]
color_dct = dict(zip(leaf_keys, color_cycle))

surf_ = mess.surf.remove_nodes(surf, [6])
node_paths = mess.surf.node_paths_from_source(surf_, root_key, leaf_keys=leaf_keys)
print(node_paths)

colors = [color_dct.get(node_path[-1], plot.Color.black) for node_path in node_paths]

node_paths = [*node_paths, [20, 6, 2], [20, 6, 16]]
colors = [*colors, plot.Color.red, plot.Color.red]

chart = mess.surf.plot_paths(
    surf,
    node_paths,
    colors=colors,
    label=True,
)

chart.properties(width=600, height=300)
```
