```python
from autochem.util import plot
import more_itertools as mit


def lighten_hex_colors(
    base_hex: str,
    n: int,
    max_lighten: float = 0.8,
) -> list[str]:
    """
    Generate n progressively lighter shades of a base color.

    Parameters
    ----------
    base_hex : str
        Base color like '#8533FF'.
    n : int
        Number of colors.
    max_lighten : float
        Maximum fraction toward white (0–1). 1 = pure white.

    Returns
    -------
    list[str]
        List of hex color strings.
    """
    base_hex = base_hex.lstrip("#")
    r = int(base_hex[0:2], 16)
    g = int(base_hex[2:4], 16)
    b = int(base_hex[4:6], 16)

    if n <= 1:
        return [f"#{r:02X}{g:02X}{b:02X}"]

    step = max_lighten / (n - 1)

    colors = []
    for i in range(n):
        t = i * step
        r_new = round(r + t * (255 - r))
        g_new = round(g + t * (255 - g))
        b_new = round(b + t * (255 - b))
        colors.append(f"#{r_new:02X}{g_new:02X}{b_new:02X}")

    return colors

root_key = 1
leaf_keys = [18, 19, 27, 17, 16, 14]
# leaf_keys = [16, 14, 17, 18, 19, 27]
color_cycle = reversed(lighten_hex_colors(plot.Color.purple, len(leaf_keys)))
color_dct = dict(zip(leaf_keys, color_cycle))

node_paths = mess.surf.node_paths_from_source(surf, root_key, leaf_keys=leaf_keys)
node_paths = sorted(node_paths, key=lambda path: mess.surf.node_object(surf, path[-1]).energy)
colors = [color_dct[path[-1]] for path in node_paths]

mess.surf.plot_paths(
    surf,
    node_paths,
    colors = colors,
    stack_nodes=[[2, 3, 4, 5]],
    label=True,
    energy_range=(-30, 40),
)
```
