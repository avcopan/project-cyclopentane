```python
import polars
import altair as alt

dataframe = polars.read_csv("AltairGraph.csv", quote_char="'")
dataframe = dataframe.with_columns(dataframe["Carbon_Count"].cast(polars.Int64))
alt.Chart(dataframe).mark_bar().encode(
    y=alt.X('count():Q', title='Number of Species'),
    x=alt.Y('Functional_Group:N', title='Primary Functional Groups'),
    color=alt.Color('Carbon_Count:N', legend=alt.Legend(title="Number of Carbons"))
)
alt.Chart(dataframe).mark_bar().encode(
    x=alt.X('count():Q', title='Species Count'),
    y=alt.Y('Functional_Group:N', title='Functional Groups'),
    color=alt.Color('Carbon_Count:N', legend=alt.Legend(title="Carbons"))
)
```
