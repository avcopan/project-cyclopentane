```python
from autochem.util import plot
import more_itertools as mit

def grayscale_hex_colors(n: int, lightest: int = 200, darkest: int = 40) -> list[str]:
    """
    Generate n evenly spaced grayscale hex colors.

    Parameters
    ----------
    n : int
        Number of colors.
    lightest : int
        Grayscale value for the lightest color (0–255).
    darkest : int
        Grayscale value for the darkest color (0–255).

    Returns
    -------
    list[str]
        List of hex color strings like '#AFAFAF'.
    """
    if n <= 1:
        val = lightest
        return [f"#{val:02X}{val:02X}{val:02X}"]

    step = (darkest - lightest) / (n - 1)

    colors = []
    for i in range(n):
        v = round(lightest + i * step)
        colors.append(f"#{v:02X}{v:02X}{v:02X}")

    return colors

root_key = 0
# leaf_keys = [13, 14, 15, 17, 18]
leaf_keys = [17, 15, 13, 14, 18]
color_cycle = [plot.Color.teal, plot.Color.cyan, plot.Color.magenta, plot.Color.lime, plot.Color.lime]
# color_cycle = grayscale_hex_colors(len(leaf_keys), lightest=180, darkest=40)
color_dct = dict(zip(leaf_keys, color_cycle))

node_paths = mess.surf.node_paths_from_source(surf, root_key, leaf_keys=leaf_keys)
node_paths = sorted(node_paths, key=lambda path: mess.surf.node_object(surf, path[-1]).energy)
colors = [color_dct[path[-1]] for path in node_paths]

mess.surf.plot_paths(
    surf,
    node_paths,
    colors = colors,
    # stack_nodes=[[1, 3, 4, 5], [6, 7, 8, 10, 11], [13, 14, 15, 17, 18]],
    stack_nodes=[[1, 3, 4, 5], [13, 14, 15, 17, 18]],
    label=True,
    energy_range=(-40, 15),
)
```
