```python
def print_arrhenius_parameters(rate, name):
    assert isinstance(rate, ac.rate.data.PlogRateFit)
    ix = next(i for i, P in enumerate(rate.Ps) if P == 1)
    k = rate.arrhenius_functions[ix]
    print(f"{name}\t{k.A:.4e}\t{k.b:.4f}\t{k.E:.2f}")

print_arrhenius_parameters(open_rxn_fit.rate, "S(1210)r0 = C5O2nexammzze = C5H6O(645)z + OH(4)\tRing-opening → OH")
print_arrhenius_parameters(e12_rxn_fit.rate, "S(1210)r0 = OH(4) + S(1288)rs0\tRing-closing (1,2) → OH")
print_arrhenius_parameters(e13_rxn_fit.rate, "S(1210)r0 = C5Oqidgnvrs + OH(4)\tRing-closing (1,3) → OH")
print_arrhenius_parameters(sciss_rxn_fit.rate, "C5H6(478) + HO2(8) = S(1210)r0\tBeta scission → HO₂")
```
