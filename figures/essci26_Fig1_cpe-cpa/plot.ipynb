```python
import altair as alt
import pandas as pd
from autochem.util import plot

# Load data
df = pd.read_csv("Experiment.csv")

# Convert to long format for Altair
df_long = df.melt(
    id_vars="temperature",
    value_vars=["cyclopentane", "cyclopentene"],
    var_name="species",
    value_name="value",
)

plot.general(
    y_data=[df["cyclopentane"], df["cyclopentene"]],
    x_data=df["temperature"],
    labels=["CPA", "CPE"],
    mark="point",
    mark_kwargs={"opacity": 1, "filled": True, "shape": "diamond", "size": 80},
    x_scale=alt.Scale(domain=(650, 1000)),
    x_label="Temperature (K)",
    y_label="Concentration (ppm)",
    colors=[plot.Color.black, plot.Color.red],
)

# # Plot
# chart = (
#     alt.Chart(df_long)
#     .mark_point(point=True)
#     .encode(
#         x=alt.X("temperature:Q", title="Temperature (K)", scale=alt.Scale(domain=(650, 1000))),
#         y=alt.Y("value:Q", title="Concentration"),
#         color=alt.Color("species:N", title="Species"),
#         tooltip=["temperature", "species", "value"]
#     )
#     .properties(width=500, height=350)
# )

# chart
```
