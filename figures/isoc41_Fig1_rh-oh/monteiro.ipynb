```python
import autochem as ac
from autochem.util import plot
import pint

R = pint.Quantity("molar_gas_constant").m_as("cal/mol/K")

r1_allyl = ac.rate.ArrheniusRateFit(order=2, A=5.98e4, b=2.561, E=-99.76 * R)
r3_alkyl = ac.rate.ArrheniusRateFit(order=2, A=2.17e5, b=2.468, E=634 * R)
r4_add = ac.rate.ArrheniusRateFit(order=2, A=3.34e18, b=-2.419, E=-7.635e2 * R)
# r4_add_highp = ac.rate.ArrheniusRateFit(order=2, A=1.52e4, b=2.307, E=-2.501e3 * R)

ac.rate.data.display(
    [r1_allyl, r3_alkyl, r4_add],
    T_range=(250, 2000),
    label=["Allyl", "Alkyl", "Add"],
    color=[plot.Color.blue, plot.Color.red, plot.Color.purple],
)
```
