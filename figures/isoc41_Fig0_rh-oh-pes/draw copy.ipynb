```python
from pathlib import Path

import automech as am
from project_utilities import p_
from protomech import mess

open_browser = True
tag = "Z_mess_v2"
mech_tag = "Z_mess_v1"
root_path = Path("../..")

stoich = "C5H9O"
mess_file_path = p_.mess_final(root_path, tag) / stoich / "mess.inp"
print(mess_file_path)

mech_json = p_.stereo_mechanism(mech_tag, "json", path=p_.data(root_path))
mech = am.io.read(mech_json)
surf = mess.surf.from_mess_input(mess_file_path)
surf = mess.surf.absorb_fake_nodes(surf, exclude_keys=[12, 20, 22, 24])

mess.net.display(surf, height="1000px", mech=mech)
```

```python
from autochem.util import plot

root_key = 19
leaf_keys = [0, 23, 25, 21, 15, 17]
colors = [plot.Color.purple, plot.Color.red, plot.Color.blue, plot.Color.orange]

node_paths = mess.surf.node_paths_from_source(surf, root_key, leaf_keys=leaf_keys)
nextra = len(node_paths) - len(colors)
colors.extend([plot.Color.brown] * nextra)

mess.surf.plot_paths(
    surf,
    node_paths,
    colors=colors,
    stack_nodes=[[20, 22, 24], [0, 23, 25, 21], [1, 3, 4]],
    stack_edges=[[(12, 20), (12, 0)]],
    label=False,
)
```
